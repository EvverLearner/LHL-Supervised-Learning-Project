```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Correlation matrix to get an overview of feature relationships
correlation_matrix = df.corr()
correlation_matrix
```

```python
# Relation of predictors to outcome
predictors = ['Age', 'BMI', 'Glucose', 'BloodPressure', 'Insulin', 'SkinThickness', 'Pregnancies', 'DiabetesPedigreeFunction']
for predictor in predictors:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Outcome', y=predictor, data=df)
    plt.title(f'{predictor} vs Outcome')
    plt.show()
```

```python
# Distribution of predictors
for column in predictors:
    plt.figure()
    sns.histplot(df[column])
    plt.title(f'Distribution of {column}')
    plt.show()
    
# Distributions also show potential outliers, especially in conjunction with practical context
```

```python
# Average age of individuals
avg_age = round(df['Age'].mean())
print(f'Average age of patients: {avg_age}')
```

```python
# Instances of diabetes positive only
diabetes_pos = df[df['Outcome'] == 1]

# Instances of diabetes negative only
diabetes_neg = df[df['Outcome'] == 0]
```

```python
# Average blood glucose and BMI for diabetic and non-diabetic individuals
avg_glucose_diabetes_pos = diabetes_pos['Glucose'].mean()
avg_glucose_diabetes_neg = diabetes_neg['Glucose'].mean()
avg_bmi_diabetes_pos = diabetes_pos['BMI'].mean()
avg_bmi_diabetes_neg = diabetes_neg['BMI'].mean()

print(f'Average Glucose (Diabetes): {avg_glucose_diabetes_pos}')
print(f'Average BMI (Diabetes): {avg_bmi_diabetes_pos}')
print('')
print(f'Average Glucose (No Diabetes): {avg_glucose_diabetes_neg}')
print(f'Average BMI (No Diabetes): {avg_bmi_diabetes_neg}')
```

```python
# Distribution of predictors by outcome, for visual comparison
for column in predictors:
    x1 = diabetes_pos[column]
    x2 = diabetes_neg[column]
    plt.figure()
    plt.hist([x1, x2])
    plt.title(f'Distribution of {column}')
    plt.legend(['Diabetic', 'Non-Diabetic'])
    plt.show()
```

# Part II : Preprocessing & Feature Engineering

You need to perform preprocessing on the given dataset. Please consider the following tasks and carry out the necessary steps accordingly.
- Handling missing values
- Handling outliers
- Scaling and normalization
- Feature Engineering
- Handling imbalanced data

```python
# Replace values of 0 that don't make sense for relevant features
zero_value_features = ['BMI', 'Glucose', 'BloodPressure', 'Insulin', 'SkinThickness']

for column in zero_value_features:
    df[column] = df[column].replace(0, df[column].mean())
    
# Check min values to see if 0s removed
df.describe()
```

```python
# Clear Outliers: skinthickness, bmi, insulin
# Dropping outliers since I'm unsure if it will bias the model
rows_to_drop = df['SkinThickness'] > 70
df_filtered = df[~rows_to_drop]

rows_to_drop = df_filtered['BMI'] > 60
df_filtered = df_filtered[~rows_to_drop]

rows_to_drop = df_filtered['Insulin'] > 700
df_filtered = df_filtered[~rows_to_drop]

# Check max values to see if outliers removed
df_filtered.describe()
```

# Part III : Training ML Model

For this task, you are required to build a machine learning model to predict the outcome variable. This will be a binary classification task, as the target variable is binary. You should select at least two models, one of which should be an ensemble model, and compare their performance.

- Train the models: Train the selected models on the training set.
- Model evaluation: Evaluate the trained models on the testing set using appropriate evaluation metrics, such as accuracy, precision, recall, F1-score, and ROC-AUC.
- Model comparison: Compare the performance of the selected models and choose the best-performing model based on the evaluation metrics. You can also perform additional analysis, such as model tuning and cross-validation, to improve the model's performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(df_filtered[predictors], df_filtered['Outcome'], test_size=0.2, random_state=3)
print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
print('')

# Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=10)
rf_model.fit(X_train, y_train)
train_score = rf_model.score(X_train, y_train)
test_score = rf_model.score(X_test, y_test)
print(f'RF Training score: {train_score}')
print(f'RF Testing score: {test_score}')
print('')

# Feature importance created by RF, sorted in descending order
feature_importance = rf_model.feature_importances_
sorted_indices = feature_importance.argsort()[::-1]
selected_features = X_train.columns[sorted_indices]
print(selected_features)
print('')

# RF with top 3 most impactful features
num_top_features = 3
selected_features = X_train.columns[sorted_indices[:num_top_features]]

rf_model_selected_features = RandomForestClassifier(n_estimators=100, random_state=10)
rf_model_selected_features.fit(X_train[selected_features], y_train)
train_score = rf_model_selected_features.score(X_train[selected_features], y_train)
test_score = rf_model_selected_features.score(X_test[selected_features], y_test)
print(f'RF Adjusted Training score: {train_score}')
print(f'RF Adjusted Testing score: {test_score}')
print('')

# F1 and ROC-AUC
y_pred = rf_model_selected_features.predict(X_test[selected_features])
y_scores = rf_model_selected_features.predict_proba(X_test[selected_features])[:, 1]
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_scores)
print(f"F1 Score: {f1:.4f}")
print(f"ROC-AUC Score: {roc_auc:.4f}")
```

```python
from sklearn.svm import SVC

# SVM with most important features from RF
svm_model = SVC(kernel='linear')
svm_model.fit(X_train[selected_features], y_train)
train_score = svm_model.score(X_train[selected_features], y_train)
test_score = svm_model.score(X_test[selected_features], y_test)
print(f'SVM Training score: {train_score}')
print(f'SVM Testing score: {test_score}')
print('')

# F1 and ROC-AUC
y_pred = svm_model.predict(X_test[selected_features])
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_scores)
print(f"F1 Score: {f1:.4f}")
print(f"ROC-AUC Score: {roc_auc:.4f}")
```
